```python
t=[]
for i in range(len(df)):
  t.append([str(df.values[i,j]) for j in range(df.shape[1]) if str(df.values[i,j]) != '0'])
```

```python
!pip install apyori
```

```python
from apyori import apriori
rules = apriori(transactions=t,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
results = list(rules)
```

```python
s =results[0][1]
```

```python
r = list(results[0][2][0])
l1 =list(r[0])[0]
l2 =list(r[1])[0]
c =r[2]
li =r[3]
print(l1,l2)
print('c',c)
print('li',li)
```

```python
def lists(results,i):
  s = results[i][1]
  r = list(results[i][2][0])
  l1 =list(r[0])[0]
  l2 =list(r[1])[0]
  c =r[2]
  li =r[3]
  return [l1,l2,s,c,li]
res = []
for i in range(len(results)):
  res.append(lists(results,i))
```

```python
pd.DataFrame(res,columns=['first','second','support','confidence','lift'])
```
